# Гипотезы

### __Гипотеза 1:__ Квартиры, которые имеют площадь кухни выше 13 кв.м. существенно дороже по цене и относятся скорее к премиальным квартирам. то есть их цена значительно превышает остальные

Рассмотрев множество графиков зависимости площади кухни и цены на квартиру пришли к предположению, что квартиры, которые имеют площадь кухни выше некоторого значения существенно дороже по цене и относятся скорее к премиальным квартирам. В качестве рассматриваемого пограничного значения взяли 90-ый перцентиль площади кухни. Сравним средние цены у квартир с площадью кухни вышего этого значения и ниже. При этом мы предположили, что цена значительно дороже, поэтому предположим, что у квартир с площадью кузнивыше 13 кв.м цена в 2 раза выше. Следовательно, переформулируем гипотезу более конкретно

__Гипотеза 1:__ Стоимость за кв.м в квартирах с большой кухней превышает 2 раза стоимость за кв.м в квартирах с маленькой кухней

Уровень значимость пусть будет $\alpha = 5\%$

##### Мат вид гипотезы:

$$
H_0: \mu_{big_kit} > 2*\mu_{lit_kit}
$$

$$
H_1: \mu_{big_kit} < 2*\mu_{lit_kit}
$$

$\mu_{big_kit}$ - средняя цена аренды за 1 кв метр в квардтирах с кухней больше 13 кв.м

$\mu_{lit_kit}$ - средняя цена аренды за 1 кв метр в квардтирах с кухней меньше 13 кв.м

```python
alpha = 0.05

big_kit = df_info[df_info['Площадь кухни'] > 13].copy()
lit_kit = df_info[df_info['Площадь кухни'] <= 13].copy()

big_kit['Стоимость за кв.м'] = big_kit['Цена'] / big_kit['Общая площадь']
lit_kit['Стоимость за кв.м'] = lit_kit['Цена'] / lit_kit['Общая площадь']


big_kit = big_kit.dropna(subset=['Стоимость за кв.м'])
lit_kit = lit_kit.dropna(subset=['Стоимость за кв.м'])

t_stat, p_value = stats.ttest_ind(big_kit['Стоимость за кв.м'], 2 * lit_kit['Стоимость за кв.м'], alternative='less')

print(f't-статистика: {t_stat}')
print(f'p-значение: {p_value}')

if p_value < alpha:
    print("Отвергаем нулевую гипотезу в пользу альтернативной")
else:
    print("Не отвергаем нулевую гипотезу")
```

В данном случае нулевая гипотеза была отвергнута, то есть стоимость за кв.м в квартирах с большой кухней не превышает 2 раза стоимость за кв.м в квартирах с маленькой кухней. Поробуем измениить коэф 2 на 1.3 и проверим гипотезу. Переформулируем гипотезу 1:

__Гипотеза 1.1:__ Стоимость за кв.м в квартирах с большой кухней превышает на 30% стоимость за кв.м в квартирах с маленькой кухней

```python
alpha = 0.05

big_kit = df_info[df_info['Площадь кухни'] > 13].copy()
lit_kit = df_info[df_info['Площадь кухни'] <= 13].copy()

big_kit['Стоимость за кв.м'] = big_kit['Цена'] / big_kit['Общая площадь']
lit_kit['Стоимость за кв.м'] = lit_kit['Цена'] / lit_kit['Общая площадь']


big_kit = big_kit.dropna(subset=['Стоимость за кв.м'])
lit_kit = lit_kit.dropna(subset=['Стоимость за кв.м'])

t_stat, p_value = stats.ttest_ind(big_kit['Стоимость за кв.м'], 1.3 * lit_kit['Стоимость за кв.м'], alternative='less')

print(f't-статистика: {t_stat}')
print(f'p-значение: {p_value}')

if p_value < alpha:
    print("Отвергаем нулевую гипотезу в пользу альтернативной")
else:
    print("Не отвергаем нулевую гипотезу")
```

__Вывод:__ гипотеза 1.1 верна, стоимость за кв.м в квартирах с большой кухней превышает на 30% стоимость за кв.м в квартирах с маленькой кухней

### __Гипотеза 2:__ всегда выгоднее снимать двушку, чем однушку

После сравнения таких переменных как 'Общая площадь', 'Отношение нежилой и жилой площадей' пришли к предположению, что выгоднее снимать двушку, чем однушку. Сравнивать будем по стоимости квадратного метра. В плане переменной 'Отношение нежилой и жилой площадей' пришли к выводу, что двушки однозначно лучше, но если рассматривать 

После разностороннего анализа квартир в зависимости от того, сколько в ней комнат, студия ли, пришли к выводу, что с точки зрения общего кол-ва квадратных метров, отношениях нежилой и жилой площадей двушки однозначно лучше. Однако, если смотреть на среднюю цену за 1 кв.м, то двушки дороже большинства квартир, но незначительно. Мы хотим с помощью  t-теста определить, ниже ли цены за 1 квадратный метр у двушек, чем у однушек.

Уровень значимость пусть будет $\alpha = 5\%$

##### Мат вид гипотезы:

$$
H_0: \mu_2 <= \mu_1
$$

$$
H_1: \mu_2 > \mu_1
$$

$\mu_{2}$ - средняя цена аренды за 1 кв метр в двушках

$\mu_{1}$ - средняя цена аренды за 1 кв метр в однушках

```python
alpha = 0.05

room_1 = df_info[df_info['Кол-во комнат'] == '1'].copy()
rooms_2 = df_info[df_info['Кол-во комнат'] == '2'].copy()

room_1['Стоимость за кв.м'] = room_1['Цена'] / room_1['Общая площадь']
rooms_2['Стоимость за кв.м'] = rooms_2['Цена'] / rooms_2['Общая площадь']

t_stat, p_value = stats.ttest_ind(room_1['Стоимость за кв.м'], rooms_2['Стоимость за кв.м'], alternative='less')

print(f't-статистика: {t_stat}')
print(f'p-значение: {p_value}')

if p_value < alpha:
    print("Отвергаем нулевую гипотезу в пользу альтернативной")
else:
    print("Не отвергаем нулевую гипотезу")
```

 __Гипотеза 3:__ Вне зависимости от года постройки монолитные дома всего ценятся выше, чем остальные.


Рассматривая переменную "Год постройки" задали график отношения типов домов в зависимости от года, пытаясь найти возможные пути объединения домой по группам (хрущевка, сталинка и тп). На этом графике видно, что понолитные дома всегда начали строить еще в 1950е, но при этом продолжали их строить не так много, достаточно равномерно, а не как панельки, в 80ые-90ие. Также мы определили, что цены на аренду квартир в монолитных домах выше большинства, поэтому свыдвинули предположение, что монолитные дома в целом более высокого класса, что делает их цены выше остальных домов все зависимости от года постройки. При этом сравнивать стоимость будем именно в цене за 1 кв. м, так как это более репрезентативно. Также важно неучитывать монолитно-кирпичные дома, так как их средняя цена значительно отличается от цены ан=ренды квартиры в моноличном доме

Уровень значимость пусть будет $\alpha = 5\%$

##### Мат вид гипотезы:

$$
H_0: \mu_{ост} < \mu_{мон}
$$

$$
H_1: \mu_{ост} >= \mu_{мон}
$$

$\mu_{ост}$ - средняя цена аренды за 1 кв метр во все домах кроме монолитного и монолино-кирпичного , построенных после 2000 года

$\mu_{мон}$ - средняя цена аренды за 1 кв метр в монолитном доме, построенном после 2000 года

```python
alpha = 0.05

monolit = df_info[df_info['Тип дома'] == 'Монолитный'].copy()
no_monolit = df_info[(df_info['Тип дома'] != 'Монолитный') & (df_info['Тип дома'] != 'Монолитно-кирпичный')].copy()

monolit['Стоимость за кв.м'] = monolit['Цена'] / monolit['Общая площадь']
no_monolit['Стоимость за кв.м'] = no_monolit['Цена'] / no_monolit['Общая площадь']

t_stat, p_value = stats.ttest_ind(monolit['Стоимость за кв.м'], no_monolit['Стоимость за кв.м'], alternative='greater')

print(f't-статистика: {t_stat}')
print(f'p-значение: {p_value}')

if p_value < alpha:
    print("Отвергаем нулевую гипотезу в пользу альтернативной")
else:
    print("Не отвергаем нулевую гипотезу")
```

Оказалось, что монолитный дом не обязательно выше по стоимости панельных, кирпичных, блочных, но гипотеза была приязана ко времени. То есть мы предполагали, что аренда в старом монолитном доме будет дороже, чем в новом панельном. Сейчас рассмотрим только относительно новые дома (были построены после 2000 года), чтобы сраdyение было более показательным

```python
alpha = 0.05

monolit = df_info[(df_info['Тип дома'] == 'Монолитный') & (df_info['Год постройки'] >= 2000)].copy()
no_monolit = df_info[(df_info['Тип дома'] != 'Монолитный') & (df_info['Тип дома'] != 'Монолитно-кирпичный') & (df_info['Год постройки'] >= 2000)].copy()

monolit['Стоимость за кв.м'] = monolit['Цена'] / monolit['Общая площадь']
no_monolit['Стоимость за кв.м'] = no_monolit['Цена'] / no_monolit['Общая площадь']


t_stat, p_value = stats.ttest_ind(monolit['Стоимость за кв.м'], no_monolit['Стоимость за кв.м'], alternative='greater')

print(f't-статистика: {t_stat}')
print(f'p-значение: {p_value}')

if p_value < alpha:
    print("Отвергаем нулевую гипотезу в пользу альтернативной")
else:
    print("Не отвергаем нулевую гипотезу")

```

Все таки отвергаем гипотезу, хотя t-статистика и p-значение стали значительно ближе друг с другом после рассмотрения домов, построенных после 2000 года. 

 __Вывод:__ Монолитные не ценятся выше, чем блочные, панельные или кирпичные

### __Гипотеза 3.1:__ кирпичные дома имеют низкую среднюю цену именно из за того, что большое количество старых кирпичных домов, которые занижают среднее, при этом цена все равно незначительно отличается от монолитного. Следовательно, цена нового кирпичного дома выше, чем цена нового монолитного.

Рассматривая переменную "Год постройки" задали график отношения типов домов в зависимости от года, пытаясь найти возможные пути объединения домой по группам (хрущевка, сталинка и тп). мы выяснили, что средние цены аренды квартир в кирпичных домах незначительно ниже цен на квартиры в монолитчных домах. При этом значительная часть кирпичных домов была построена до 1966 года, что делает эти дома менее ценными из за их возраста. Сейчас проверим, правда ли аренда квартир в кирпичных домах, построенных после 2000 года выше, чем в монолитных домах, построенныъ тоже построе 2000 года. При этом сравнивать стоимость будем именно в цене за 1 кв. м, так как это более репрезентативно.

Уровень значимость пусть будет $\alpha = 5\%$

##### Мат вид гипотезы:

$$
H_0: \mu_{кирп} > \mu_{мон}
$$

$$
H_1: \mu_{кирп} <= \mu_{мон}
$$

$\mu_{кирп}$ - средняя цена аренды за 1 кв метр в кирпичном доме, построенном после 2000 года

$\mu_{мон}$ - средняя цена аренды за 1 кв метр в монолитном доме, построенном после 2000 года

```python
alpha = 0.05

monolit = df_info[(df_info['Тип дома'] == 'Монолитный') & (df_info['Год постройки'] >= 2000)].copy()
kirpich = df_info[(df_info['Тип дома'] == 'Кирпичный') & (df_info['Год постройки'] >= 2000)].copy()

monolit['Стоимость за кв.м'] = monolit['Цена'] / monolit['Общая площадь']
kirpich['Стоимость за кв.м'] = kirpich['Цена'] / kirpich['Общая площадь']

t_stat, p_value = stats.ttest_ind(monolit['Стоимость за кв.м'], kirpich['Стоимость за кв.м'], alternative='greater')

print(f't-статистика: {t_stat}')
print(f'p-значение: {p_value}')

if p_value < alpha:
    print("Отвергаем нулевую гипотезу в пользу альтернативной")
else:
    print("Не отвергаем нулевую гипотезу")
```

__Вывод:__ кирпичные дома и правда дороже монолитных, если рассматривать дома одного периода времени, в частности, простроенные после 2000 года
