```python
from instagram_private_api import Client
from instagram_private_api.errors import ClientError

# Функция для получения статистики по рилсам пользователя
def get_rails_stats(username):
    try:
        # Создание клиента Instagram
        client = Client()

        # Получение информации о пользователе по его нику
        user_info = client.username_info(username)
        user_id = user_info['user']['pk']

        # Получение списка постов пользователя
        posts = client.user_feed(user_id)

        # Словарь для хранения статистики по рилсам
        rails_stats = {}

        # Перебор постов и сбор статистики по рилсам
        for post in posts['items']:
            if post['media_type'] == 2 and 'view_count' in post['video_versions'][0]:
                shortcode = post['code']
                views = post['video_view_count']
                likes = post['like_count']
                rails_stats[shortcode] = {'views': views, 'likes': likes}

        return rails_stats

    except ClientError as e:
        print("Произошла ошибка при работе с Instagram API:", e)
        return None

# Указание ника блогера
username = 'mayskaya_a_a'

# Получение статистики по рилсам блогера
rails_stats = get_rails_stats(username)

# Вывод результатов
if rails_stats is not None:
    print("Статистика по рилсам пользователя:")
    for shortcode, stats in rails_stats.items():
        print(f"Пост: {shortcode}")
        print(f"Просмотры: {stats['views']}")
        print(f"Лайки: {stats['likes']}")
        print()
else:
    print("Не удалось получить статистику по рилсам.")
```
