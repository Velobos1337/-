```python
from instagram_private_api import Client
from instagram_private_api.errors import ClientError

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –ø–æ —Ä–∏–ª—Å–∞–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
def get_rails_stats(username):
    try:
        # –°–æ–∑–¥–∞–Ω–∏–µ –∫–ª–∏–µ–Ω—Ç–∞ Instagram
        client = Client()

        # –ü–æ–ª—É—á–µ–Ω–∏–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ –ø–æ –µ–≥–æ –Ω–∏–∫—É
        user_info = client.username_info(username)
        user_id = user_info['user']['pk']

        # –ü–æ–ª—É—á–µ–Ω–∏–µ —Å–ø–∏—Å–∫–∞ –ø–æ—Å—Ç–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
        posts = client.user_feed(user_id)

        # –°–ª–æ–≤–∞—Ä—å –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –ø–æ —Ä–∏–ª—Å–∞–º
        rails_stats = {}

        # –ü–µ—Ä–µ–±–æ—Ä –ø–æ—Å—Ç–æ–≤ –∏ —Å–±–æ—Ä —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –ø–æ —Ä–∏–ª—Å–∞–º
        for post in posts['items']:
            if post['media_type'] == 2 and 'view_count' in post['video_versions'][0]:
                shortcode = post['code']
                views = post['video_view_count']
                likes = post['like_count']
                rails_stats[shortcode] = {'views': views, 'likes': likes}

        return rails_stats

    except ClientError as e:
        print("–ü—Ä–æ–∏–∑–æ—à–ª–∞ –æ—à–∏–±–∫–∞ –ø—Ä–∏ —Ä–∞–±–æ—Ç–µ —Å Instagram API:", e)
        return None

# –£–∫–∞–∑–∞–Ω–∏–µ –Ω–∏–∫–∞ –±–ª–æ–≥–µ—Ä–∞
username = 'mayskaya_a_a'

# –ü–æ–ª—É—á–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –ø–æ —Ä–∏–ª—Å–∞–º –±–ª–æ–≥–µ—Ä–∞
rails_stats = get_rails_stats(username)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
if rails_stats is not None:
    print("–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ —Ä–∏–ª—Å–∞–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è:")
    for shortcode, stats in rails_stats.items():
        print(f"–ü–æ—Å—Ç: {shortcode}")
        print(f"–ü—Ä–æ—Å–º–æ—Ç—Ä—ã: {stats['views']}")
        print(f"–õ–∞–π–∫–∏: {stats['likes']}")
        print()
else:
    print("–ù–µ —É–¥–∞–ª–æ—Å—å –ø–æ–ª—É—á–∏—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –ø–æ —Ä–∏–ª—Å–∞–º.")
```
